```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='numShots', y='Shot Zone Range', data=shotDistances, palette='viridis')

# Add labels and title
bar_plot.set_xlabel('Number of Shots')
bar_plot.set_ylabel('Shot Zone Range')
bar_plot.set_title('Number of Shots by Shot Zone Range')

# Add values on the bars
for index, value in enumerate(shotDistances['numShots']):
    bar_plot.text(value, index, f' {value}')

# Display the plot
plt.show()
```

```python
# we're gonna remove back court shots since those are mostly end of quarter heaves.
shotDistances = shotDistances[shotDistances['Shot Zone Range'] != 'Back Court Shot']

import pandas as pd
import matplotlib.pyplot as plt

# Create the pie chart
plt.figure(figsize=(10, 6))
colors = ['yellow', 'orange', 'red', 'teal']
plt.pie(shotDistances['numShots'], labels=shotDistances['Shot Zone Range'], autopct='%1.1f%%', startangle=140, colors=colors)

# Add a title
plt.title('Proportion of Shots by Shot Zone Range')

# Display the plot
plt.show()
```

```python
seasonFirst = 19970801
seasonLast = 20190801

# assume august 1 because it's the offseason
```

```python
fSeasonData = full_data[full_data['Game Date'] > seasonFirst]
fSeasonData = fSeasonData[fSeasonData['Game Date'] < (seasonFirst + 10000)]
fSeasonData = fSeasonData.sort_values(by='Game Date', ascending=True)
fSeasonData
```

```python
lSeasonData = full_data[full_data['Game Date'] > seasonLast]
lSeasonData = lSeasonData.sort_values(by='Game Date', ascending=True)
lSeasonData
```

```python
def plotPieChart(seasonData, seasonNumber):
    shotDistances = seasonData.groupby('Shot Zone Range').size().reset_index(name='numShots')
    
    # we're gonna remove back court shots since those are mostly end of quarter heaves.
    shotDistances = shotDistances[shotDistances['Shot Zone Range'] != 'Back Court Shot']

    # Create the pie chart
    plt.figure(figsize=(10, 6))
    colors = ['yellow', 'orange', 'red', 'teal']
    plt.pie(shotDistances['numShots'], labels=shotDistances['Shot Zone Range'], autopct='%1.1f%%', startangle=140, colors=colors)

    # Add a title
    plt.title('Proportion of Shots by Shot Zone Range in ' + seasonNumber)

    # Display the plot
    plt.show()
```

```python
plotPieChart(fSeasonData, '1997/98')
plotPieChart(lSeasonData, '2019/20')
```

```python
import pandas as pd

# Convert 'Game Date' to datetime format
full_data['Game Date'] = pd.to_datetime(full_data['Game Date'], format='%Y%m%d')

# Function to determine the season
def get_season(date):
    year = date.year
    month = date.month
    
    if month >= 8:  # August or later
        season_start = year
        season_end = year + 1
    else:  # Before August
        season_start = year - 1
        season_end = year
    
    return f'{season_start}-{str(season_end)[-2:]}'

# Apply the function to determine the season for each 'Game Date'
full_data['Season'] = full_data['Game Date'].apply(get_season)

# Display the DataFrame
# print(full_data)
full_data
```

```python
full_data.sort_values(by='Season', ascending=False)
```

```python
nearest = full_data[full_data['Shot Zone Range'] != 'Less Than 8 ft.']
nearest = nearest[nearest['Shot Zone Range'] != '24+ ft.']
nearest
```

```python
furthest = full_data[full_data['Shot Zone Range'] == '24+ ft.']
furthest
```

```python
midrange = nearest.groupby('Season').size().reset_index(name='shot count')
```

```python
midrange
```

```python
threes = furthest.groupby('Season').size().reset_index(name='shot count')
```

```python
threes
```

```python
# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting bars for 'paint'
ax.bar(midrange['Season'], midrange['shot count'], width=0.4, label='Midrange', align='center', color='red')

# Plotting bars for 'threes' with an offset to separate the bars
ax.bar(threes['Season'], threes['shot count'], width=0.4, label='Threes', align='edge', color='blue')

# Adding labels and title
ax.set_xlabel('Season')
ax.set_ylabel('Shot Count')
ax.set_title('Comparison of midrange shots vs three point shots')

# Adding legend
ax.legend()

# Display the plot
plt.tight_layout()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
allShots = full_data.groupby('Season').size().reset_index(name='total shot count')
midrangeShots = nearest.groupby('Season').size().reset_index(name='midrange shot count')
threePtShots = furthest.groupby('Season').size().reset_index(name='three shot count')
```

```python
# Merge allShots and midrangeShots on 'Season'
merged_df = pd.merge(allShots, midrangeShots, on='Season')

# Merge merged_df with threePtShots on 'Season'
merged_df = pd.merge(merged_df, threePtShots, on='Season')

# Display the merged DataFrame
merged_df
```

```python
merged_df['midrange%'] = merged_df['midrange shot count'] /  merged_df['total shot count'] * 100
merged_df['three%'] = merged_df['three shot count'] /  merged_df['total shot count'] * 100
```

```python
# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting bars for 'paint'
ax.bar(merged_df['Season'], merged_df['midrange%'], width=0.4, label='Midrange', align='center', color='red')

# Plotting bars for 'threes' with an offset to separate the bars
ax.bar(merged_df['Season'], merged_df['three%'], width=0.4, label='Threes', align='edge', color='blue')

# Adding labels and title
ax.set_xlabel('Season')
ax.set_ylabel('% of total shots')
ax.set_title('Comparison of midrange shot % vs three point shot %')

# Adding legend
ax.legend()

# Display the plot
plt.tight_layout()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```
