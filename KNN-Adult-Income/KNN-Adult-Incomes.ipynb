```python
import numpy as np
import math

def knn(adultInfo, features, labels, k):
    predictedLabels = []
    distances = []
    for i in range(len(features)):
        dist = features[i] - adultInfo
        dist_val = math.sqrt(np.dot(dist, dist))
        distances.append((dist_val, labels[i]))
    
    dtype = [('distance', float), ('label', int)]
    distances = np.array(distances, dtype=dtype)
    distances = np.sort(distances, order=['distance'])
    
    for i in range(k):
        predictedLabels.append(distances[i]['label'])

    return np.array(predictedLabels)

def classify_devset(dev_data, train_data, train_labels, k):
    '''
    Parameters:
    dev_images (list) -M images
    train_images (list) -N images
    train_labels (list) -N labels corresponding to the N images
    k (int) - the number of neighbors to use for each dev image

    Output:
    hypotheses (list) -one majority-vote labels for each of the M dev images
    scores (list) -number of nearest neighbors that voted for the majority class of each dev image
    '''
    hypotheses = []
    scores = []
    for data in dev_data:
        labels = knn(data, train_data, train_labels, k)
        num_zeros = 0
        num_ones = 0
        for i in labels:
            if i == 0:
                num_zeros += 1
            else:
                num_ones += 1
        if num_zeros >= num_ones:
            hypotheses.append(0)
            scores.append(num_zeros)
        else:
            hypotheses.append(1)
            scores.append(num_ones)
    return np.array(hypotheses), np.array(scores)

def confusion_matrix(hypotheses, references):
    '''
    Parameters:
    hypotheses (list) - a list of M labels output by the classifier
    references (list) - a list of the M correct labels

    Output:
    confusions (list of lists, or 2d array) - confusions[m][n] is 
    the number of times reference class m was classified as
    hypothesis class n.
    accuracy (float) - the computed accuracy
    f1(float) - the computed f1 score from the matrix
    '''
    tp = 0
    fp = 0
    tn = 0
    fn = 0
    for i in range(len(hypotheses)):
        if hypotheses[i] == 0:
            if references[i] == 0:
                tn += 1
            else:
                fn += 1
        else:
            if references[i] == 0:
                fp += 1
            else:
                tp += 1

    confusions = [[tn, fp], [fn, tp]]
    confusions = np.array(confusions)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    accuracy = (tp + tn) / (tp + tn + fn + fp)
    f1 = 2 / ((1 / recall) + (1 / precision))
    
    return confusions, accuracy, f1
```
