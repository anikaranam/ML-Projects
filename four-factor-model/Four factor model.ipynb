```python
import pandas as pd
```

```python
data = pd.read_csv('~/Desktop/2017-22_complete.csv')
```

```python
data
```

```python
data = data[data['MATCH UP'] != 'MATCH UP']
```

```python
data
```

```python
def extract_first_team(matchup):
    return matchup.split()[0]

# Apply the function to create a new column 'Team'
data['Team'] = data['MATCH UP'].apply(extract_first_team)
```

```python
data
```

```python
mean_values = data.groupby('Team').agg({
    'eFG%': 'mean',
    'FTARate': 'mean',
    'TOV%': 'mean',
    'OREB%': 'mean',
    'OppFTA': 'mean',
    'OppeFG%': 'mean',
    'OppTOV%': 'mean',
    'OppOREB%': 'mean'
}).reset_index()

mean_values
```

```python
data.columns
```

```python
data_types = data.dtypes

data_types
```

```python
columns_to_convert = ['eFG%', 'FTARate', 'TOV%', 'OREB%', 'OppeFG%', 'OppFTA', 'OppTOV%', 'OppOREB%']

for column in columns_to_convert:
    data[column] = pd.to_numeric(data[column], errors='coerce')
```

```python
def getResult(winloss):
    if winloss == 'W':
        return 1
    return 0

data['win'] = data['W/L'].apply(getResult)

data
```

```python
mean_values = data.groupby('Team').agg({
    'eFG%': 'mean',
    'FTARate': 'mean',
    'TOV%': 'mean',
    'OREB%': 'mean',
    'OppeFG%': 'mean',
    'OppFTA': 'mean',
    'OppTOV%': 'mean',
    'OppOREB%': 'mean',
    'win': 'sum'
}).reset_index()

mean_values
```

```python
mean_values = mean_values[mean_values['win'] > 0]
```

```python
mean_values
```

```python
len(mean_values)
```

```python
mean_values['win'] = mean_values['win'] / 4
```

```python
mean_values
```

```python
vals = mean_values
```

```python
vals
```

```python
vals['fgdiff'] = vals['eFG%'] - vals['OppeFG%']
vals['tovdiff'] = vals['TOV%'] - vals['OppTOV%']
vals['ftdiff'] = vals['FTARate'] - vals['OppFTA']
vals['rebdiff'] = vals['OREB%'] - vals['OppOREB%']
```

```python
vals
```

```python
vals.columns
```

```python
x = vals[['fgdiff', 'tovdiff', 'ftdiff', 'rebdiff']]
y = vals['win']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
```

```python
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))

# Display model coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
